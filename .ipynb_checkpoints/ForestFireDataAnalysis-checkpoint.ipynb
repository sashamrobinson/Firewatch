```python
# Data handling
import numpy as np
import pandas as pd

# Data visualization
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# Citation:
# P. Cortez and A. Morais. A Data Mining Approach to Predict Forest Fires using Meteorological Data.
# In J. Neves, M. F. Santos and J. Machado Eds., New Trends in Artificial Intelligence,
# Proceedings of the 13th EPIA 2007 - Portuguese Conference on Artificial Intelligence, December,
# Guimaraes, Portugal, pp. 512-523, 2007. APPIA, ISBN-13 978-989-95618-0-9.
data = pd.read_csv("forestfires.csv")
```

```python
data.describe()
```

```python
# Reference image
img = mpimg.imread("Montesinho Natural Park Nov 2023 .png")

# Subplots
fig, ax = plt.subplots()

ax.imshow(img)

grid_rows = 9
grid_cols = 9

# Determine the width and height of each cell in grid
cell_width = img.shape[1] / grid_cols
cell_height = img.shape[0] / grid_rows

# Draw grid lines
for i in range(grid_cols + 1):
    x = i * cell_width
    ax.axvline(x, color='black', linewidth=0.5)

for i in range(grid_rows + 1):
    y = i * cell_height
    ax.axhline(y, color='black', linewidth=0.5)

# Set the axis limits to match the image size
ax.set_xlim(0, img.shape[1])
ax.set_ylim(img.shape[0], 0)

# Set axis ticks
ax.set_xticks([cell_width / 2 + i * cell_width for i in range(grid_cols)])
ax.set_xticklabels([str(i + 1) for i in range(grid_cols)])
ax.set_yticks([cell_height / 2 + i * cell_height for i in range(grid_rows)])
ax.set_yticklabels([str(i + 1) for i in range(grid_rows)])

# Set axis labels
ax.tick_params(top=True, labeltop=True, bottom=False, labelbottom=False)
ax.xaxis.set_label_position('top') 
ax.set_title('Montesinho Natural Park - Reference Image')
ax.set_xlabel('X Coordinates')
ax.set_ylabel('Y Coordinates')

plt.show()
```

```python
# Get information from data set
x_coordinates = data["X"]
y_coordinates = data["Y"]
temp_values = data["temp"]
wind_values = data["wind"]
rain_values = data["rain"]

ffmc_values = data["FFMC"]

# Package data into pivot table
heatmap1_data = data.pivot_table(values='(C°)', index='Y', columns='X')
heatmap2_data = data.pivot_table(values='(km/h)', index='Y', columns='X')
heatmap3_data = data.pivot_table(values='(mm/m2)', index='Y', columns='X')
heatmap4_data = data.pivot_table(values='(%)', index='Y', columns='X')


# Prepare subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 14))
ax1, ax2, ax3, ax4 = axs.flatten()

# Temperature Heatmap
ax1.tick_params(top=True, labeltop=True, bottom=False, labelbottom=False)
ax1.xaxis.set_label_position('top') 
ax1.set_xlabel('X Coordinates')
ax1.set_ylabel('Y Coordinates')
ax1.set_title('Temperature Heatmap')
sns.heatmap(heatmap_data, cmap='Oranges', annot=True, fmt=".2f", cbar_kws={'label': 'FFMC'}, ax=ax1)

# Wind Heatmap
ax2.tick_params(top=True, labeltop=True, bottom=False, labelbottom=False)
ax2.xaxis.set_label_position('top') 
ax2.set_xlabel('X Coordinates')
ax2.set_ylabel('Y Coordinates')
ax2.set_title('Wind Speed Heatmap')
sns.heatmap(heatmap_data, cmap='Oranges', annot=True, fmt=".2f", cbar_kws={'label': 'FFMC'}, ax=ax2)

# Rain Heatmap
ax1.tick_params(top=True, labeltop=True, bottom=False, labelbottom=False)
ax1.xaxis.set_label_position('top') 
ax1.set_xlabel('X Coordinates')
ax1.set_ylabel('Y Coordinates')
ax1.set_title('Outside Rain Heatmap')
sns.heatmap(heatmap_data, cmap='Oranges', annot=True, fmt=".2f", cbar_kws={'label': 'FFMC'}, ax=ax1)

# Humidity Heatmap
ax1.tick_params(top=True, labeltop=True, bottom=False, labelbottom=False)
ax1.xaxis.set_label_position('top') 
ax1.set_xlabel('X Coordinates')
ax1.set_ylabel('Y Coordinates')
ax1.set_title('Relative Heatmap')
sns.heatmap(heatmap_data, cmap='Oranges', annot=True, fmt=".2f", cbar_kws={'label': 'FFMC'}, ax=ax1)

plt.show()
```
